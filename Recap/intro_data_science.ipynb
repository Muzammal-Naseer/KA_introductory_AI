### Download the data

```python
!wget https://raw.githubusercontent.com/selva86/datasets/master/ChickWeight.csv

data = pd.read_csv('ChickWeight.csv')
```

### Exploratory Data Analysis (EDA)

```python
# examine the data
data.head()
```

```python
# descriptive stats
display(data.describe())
```

ðŸ”‘: Use the `.describe()` function to display a statistical overview of the data.

```python
# displaying unique values in the data for each column
display(data.nunique())
```

ðŸ”‘: Use the `.nunique()` to display the number of unique values in each column.

Since we have 4 unique values for the diet, we can visualise the data
by grouping all the chicks in their respected diet categories

```python
len(data)
```

```python
display(data.pivot_table(columns=['Diet'], aggfunc='nunique'))
```

ðŸ‘‹: We created a new table from the data based on the criteria we want. We wanted
to see the chicks in each diet (unique ones, no duplicates)

To better visualise what we are aiming to do, we can create a plot of it:


```python
# get the counts in each category (Diet)
table = data.pivot_table(columns=['Diet'], aggfunc='nunique')
counts = table.loc['Chick']
# create a bar chart for the values
plt.bar(counts.index, counts.values)
# plot the data
plt.xlabel('Diet')
plt.ylabel('Count')
plt.show()
```

We can conclude the following from the observations made:

- The chicken group on diet 1 seem to be averagely fatter

- The diet 1 have 20 chickens while the other have only 10

Lets now compare the effect of the diets, ie. before (min. time) and after (max. time)

```python
before_data = data[data['Time'] == data['Time'].min()]
after_data = data[data['Time'] == data['Time'].max()]

print(f'Before diet: There were {len(before_data)} chicks')
print(f'After diet: There were {len(after_data)} chicks')
```

We can clearly see that the number of chicks at the beginning is not equal
to the number of chicks at the end. What happened to them?

```python
# instead of using matplotlib, we can use seaborn library
plt.title('Before the Diet')
sns.countplot(data=before_data, x='Diet');
```

```python
plt.title('After the Diet')
sns.countplot(data=after_data, x='Diet');
```

We can see that 4 chicks are missing from Diet 1, in addition to 1 from Diet 4.

```python
# lets group the data by the unique chicks and find the ones that didnt make it
data_chick = data.groupby('Chick')
data_chick.describe()
```

We can visually from the table see the chicks that didnt make it, however, we
can write a line of code that returns to us the abnormal chicks:

```python
# every chick complete 12 days with the diet, so the ones that didnt make it
# are what we are looking for.

MASK = np.array(data_chick.describe()['Diet']['count'] == 12)
chicks = data_chick.describe().index
missing_chicks = chicks[~MASK]
missing_chicks.values
```

Yay! We found the missing chicks. We can remove them and ensure consistent findings.
